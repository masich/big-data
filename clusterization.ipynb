```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder

%matplotlib inline 
```

Read the data.

```python
df = pd.read_csv('data/ads.csv')
df.head()
```

Label encoding to convert the data.

```python
le = LabelEncoder()

df['Time'] = le.fit_transform(df['Time'])

X = df.values
X[:5]
```

Clusterization.

In this test we will use suck clusterizers as:
1. KMeans with 2 clusters
- KMeans with 5 clusters
- AgglomerativeClustering with 5 clusters
- AgglomerativeClustering with 2 clusters

```python
clusterizers = (
    KMeans(n_clusters=5),
    KMeans(n_clusters=2),
    AgglomerativeClustering(n_clusters=5),
    AgglomerativeClustering(n_clusters=2),
)
```

```python
def perform_clustering() -> None:
    for clusterizer in clusterizers:
        # Fit the data and predict final labels
        labels = clusterizer.fit_predict(X)
        # Prefill plt by data
        plt.title(clusterizer)
        plt.scatter(X[:, 0], X[:, 1], c=labels, s=25)
        plt.show()
```

```python
perform_clustering()
```
