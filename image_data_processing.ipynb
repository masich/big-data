```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
```

Read the image data from the file.

```python
pic = imageio.imread('data/animal.jpg', as_gray=False, pilmode='RGB')
```

Show the original image.

```python
plt.imshow(pic)
plt.show()
```

Prepare the image data for processing.

```python
height, width, pixels = np.shape(pic)

picture = np.array(pic, dtype='int64')
picture = picture.reshape((height * width, pixels))
```

Define function to process pixels data.

```python
def formulae(pixel: np.ndarray) -> np.ndarray:
    r, g, b = pixel
    result = 255 if g - b >= 20 else 0
    return np.array([result] * 3)
```

```python
picture = np.array(list(map(formulae, picture)))
picture = picture.reshape((height, width, pixels))
```

Show the processed image.

```python
plt.imshow(picture)
plt.show()
```
